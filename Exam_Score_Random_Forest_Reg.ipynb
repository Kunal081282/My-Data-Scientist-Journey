```python
course = {'diploma':1, 'bca':2, 'b.sc':3, 'b.tech':4, 'bba':5, 'ba':6, 'b.com':7}
df['course'] = df['course'].map(course)
df.head()
```

```python
df['internet_access'].unique()
```

```python
internet = {'yes':1, 'no':0}
df['internet_access'] = df['internet_access'].map(internet)
df.head()
```

```python
df['sleep_quality'].unique()
```

```python
sleep = {'poor':1, 'average':2, 'good':3}
df['sleep_quality'] = df['sleep_quality'].map(sleep)
df.head()
```

```python
df['study_method'].unique()
```

```python
study = {'coaching':1, 'online videos':2, 'mixed':3, 'self-study':4, 'group study':5}
df['study_method'] = df['study_method'].map(study)
df.head()
```

```python
df['facility_rating'].unique()
```

```python
facility = {'low':1, 'medium':2, 'high':3}
df['facility_rating'] = df['facility_rating'].map(facility)
df.head()
```

```python
df['exam_difficulty'].unique()
```

```python
exam_difficulty = {'hard':3, 'moderate':2, 'easy':1}
df['exam_difficulty'] = df['exam_difficulty'].map(exam_difficulty)
df.head()
```

```python
gender = {'male':1, 'female':2, 'other':3}
df['gender'] = df['gender'].map(gender)
df.head()
```

```python
X = df[['age', 'gender', 'course', 'study_hours','class_attendance', 'internet_access', 'sleep_hours', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty']]
y = df['exam_score']
```

```python
print(X)
```

```python
print(y)
```

```python
## Splitting Data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
## Lets preprocess the data to prevent the leakage. Also the test data will remain unseen during training
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
num_features = X.columns
preprocessor = ColumnTransformer([("num", StandardScaler(), num_features)])
```

```python
##fit the model. In this case we will use logistics regression
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)
```

```python
##Train the data within a pipeline.
from sklearn.pipeline import Pipeline
pipeline = Pipeline([("preprocessing", preprocessor), ("model", model)])
pipeline.fit(X_train, y_train)
```

```python
##Evaluate - check metrics, accuracy score etc.
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y_pred = pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("RMSE:", rmse)
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
#hyperparameters tuning and utilizing Grid Search CV to find out the best parameter
from sklearn.model_selection import GridSearchCV
param_grid = {
    "model__n_estimators": [100, 200, 300],
    "model__max_depth": [None, 10, 20, 30],
    "model__min_samples_split": [2, 5, 10],
    "model__min_samples_leaf": [1, 2, 4],
    "model__max_features": ["sqrt", "log2"]
}
```

```python
grid = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    scoring="neg_root_mean_squared_error",
    cv=5,
    n_jobs=-1
)

grid.fit(X_train, y_train)
```

```python
##finding the best parameter after tuning
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
print(y_pred)
```

```python
print("RMSE:", rmse)
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
print("Best Parameters:", grid.best_params_)
```
